#**  Power BI Visuals**

                                              
Question 1 : Create a Card Visual to show Total Sales.

Ans: Create a Card visual and drag the Sales field into it.
 The card will display the Total Sales value of the dataset (sum of Sales).


Question 2  : Create a Donut Chart showing Total Sales by Ship Mode and which ship mode generates the highest sales?

Ans: The donut chart shows the contribution of each Ship Mode to total sales.
Usually, Standard Class generates the highest sales, followed by Second Class, First Class, and Same Day.

Question 3 : Create a Line Chart showing the Sales Trend over time (by Month and Year). Is there any noticeable seasonal pattern in sales?

Ans: The line chart shows the monthly sales trend.
There is usually a seasonal pattern, with higher sales toward the end of the year (Nov–Dec) due to holiday shopping.

Question 4 : Create a Bar Chart to compare Sales by Category.

Ans: The bar chart compares sales across categories.
 Typically:
Technology has the highest sales


Followed by Furniture


Then Office Supplies


Question 5 : Create a Matrix Visual to display Sales by Month and Category and identify the month with the highest sales.

Ans: The matrix helps identify which month has the highest sales.
Usually, November or December shows the highest sales due to seasonal demand.

Question 6 : Create a Decomposition Tree for Sales Breakdown.

Ans: The decomposition tree breaks down Sales into different dimensions (Category, Region, Segment, etc.), helping identify which factors contribute most to total sales.

Question 7 : Create a Scatter Chart to compare Discount vs. Profit by Sub-Category. Are there any outliers where high discounts resulted in negative profit?

Ans: The scatter chart shows the relationship between discount and profit.
Yes, there are outliers where high discounts lead to negative profit, especially in some sub-categories like Tables or Bookcases.

Question 8 : Use a Funnel Chart to show the number of orders by Customer Segment.

Ans:  The funnel chart shows the number of orders by segment.
 Typically:
Consumer – Highest


Corporate


Home Office – Lowest




Question 9 : Use a Waterfall Chart to analyze how Profit changes by Category within the countries.

Ans: The waterfall chart shows how each category contributes to total profit across countries, highlighting positive and negative profit contributions.

Question 10 : Create a Dual-Axis Line Chart to show Total Sales and Profit Over Time (by Month).

Ans: The dual-axis chart compares Sales and Profit trends over time, helping identify whether profit increases when sales increase and detect periods where sales were high but profit was low.
